```python
#plotting the count of sector kiva loans sanctioned
plt.plot(figsize=(100,100))
plt.xticks(rotation='vertical')
sns.countplot(x='country',data=kiva_loan_data)
# plt.figure(figsize=(12,6))
```

```python
# top country count
topcountry=pd.DataFrame(kiva_loan_data.country.value_counts().head(20))
topcountry=topcountry.reset_index()
topcountry.columns=['country','Count']
topcountry
```

```python
plt.plot(figsize=(15,5))
plt.xticks(rotation='vertical')
sns.barplot(x='country',y='Count',data=topcountry)
```

```python
sns.regplot(x=kiva_loan_data["funded_amount"], y=kiva_loan_data["loan_amount"], fit_reg=True)
```

```python
sns.boxplot(data=kiva_loan_data[['funded_amount','loan_amount']])
```

```python
plt.xticks(rotation='vertical')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.00), ncol=2)
sns.countplot(x='sector',data=kiva_loan_data,hue='repayment_interval')
```

```python
# joint plot

sns.jointplot("funded_amount", "loan_amount", kiva_loan_data, kind="reg")
```

```python
# correlation
loan_num = kiva_loan_data[['funded_amount','loan_amount']]
correlations = loan_num.corr()

fig, ax=plt.subplots()
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
ax.set(xticklabels=loan_num.columns,yticklabels=loan_num.columns)
plt.show()
```

```python
# heatmap of correlations
sns.heatmap(correlations)
```

```python
fig, ax = plt.subplots() #boxplot
ax.boxplot("loan_amount", data=kiva_loan_data)
plt.show()
```

```python

#scatter - point plot to understand spread of data
colors = ("red", "green")
groups = ("funded_amount", "loan_amount")
fig, ax=plt.subplots()
ax.set(title="scatterplot of funded_amount and loan_amount", xlabel="funded_amount", ylabel="loan_amount")
ax.scatter('funded_amount', 'loan_amount', alpha=0.8, c=colors, label=groups, data=kiva_loan_data) 
plt.show()
```
